````python
import re
def pass_n_fail(pattern, pass_list, fail_list):
    for p in pass_list:
        if not re.search(pattern, p):
            print("ERRORE: '{}' non matcha il pattern '{}' ma dovrebbe farlo!".format(p, pattern))
    for f in fail_list:
        if re.search(pattern, f):
            print("ERRORE: '{}' matcha il pattern '{}' ma non dovrebbe farlo!".format(f, pattern))

pass_n_fail('c', ['aa', 'a'], ['b', 'c'])
print()
pass_n_fail('a', ['aa', 'a'], ['b','c'])
print()
pass_n_fail('a', ['ab', 'ba'],['fd','zb'])
print()
pass_n_fail('a', ['a', '(aa)+b'], ['c', 'bb'])
print()
pass_n_fail('ab+', ['abc','abe'], ['b','c'])
print()
pass_n_fail('ab+', ['abb', 'abbb'], ['bbb', 'aa'])
print()
pass_n_fail('ab*',['a','ab'],['dd', 'tt'])
print()
pass_n_fail('^[ab]+[^ab]$',['bc','aaac'],['acccca','cbac'])
print()
pass_n_fail('.?\.{3}$',['...','1...'],['bar.','..'])
````

````python
import ipywidgets as widgets
from ipywidgets import Button, HBox, VBox, Tab, IntSlider, Label, HTML, AppLayout, Layout

# !!! IMPORTANTE !!!
# Il 'pyplot' che vediamo qui sotto, che viene importato con il nome di 'plt'
# proviene dalla libreria di bqplot, NON è lo stesso pyplot di matplotlib !!
# Gli autori di bqplot hanno adottato lo stesso nome e convenzioni per permetterci
# di riusare facilmente esempi che già conosciamo di matplotlib

from bqplot import pyplot as plt

x = [2,4,6]
fig = plt.figure() # genera la figure

lines = plt.plot(x, [15, 3, 20])
plt.title('Grafico in bqplot')

slider1 = IntSlider()

bottone_vai_pag2 = Button(description='VAI PAGINA 2')

slider2 = IntSlider()
hbox2 = HBox([Button(description='clicca qui'), Button(description='cliccami!')])

tab1 = HBox(children = [fig, VBox([slider1, bottone_vai_pag2])])
tab2 = VBox(children = [slider2, hbox2])

# al momento la prima 'pagina' è il widget Tab
pagina1 = widgets.Tab(children = [tab1, tab2], layout=Layout(min_height='350px'))

pagina1.set_title(0, 'TAB COL PLOT')
pagina1.set_title(1, 'ALTRA TAB')

bottone_vai_pag1 = Button(description = 'VAI PAGINA 1')

pagina2 = HBox([Label('Questa è la seconda pagina'),bottone_vai_pag1], layout = Layout(min_height='350px'))

# Il codice HTML è il codice con cui sono scritte le pagine web, qui lo 
# usiamo per creare il titolo come esempio ma non è indispensabile conoscerlo

titolo = HTML('<h1 style="color:orange">Webapp Incredibile</h1> <br/>')

# testo comune in fondo alla pagina
credits = Label('Credits: Interfacce Incredibili SRL')

# La struttura della nostra webapp è una pila VBox di elementi
my_app = VBox(children = [titolo, # supponiamo cche il titolo sia sempre visibile in tutto il sito
                           pagina1, # al momanto la prima 'pagina' è il widget tab
                             credits]) # supponiamo che il titolo sio sempre visibile in tutto il sito

# questa funzione permettew di cambiare la parte centrale della webapp passando un nuovo widget

def cambia_pagina(nuova_pagina):
    #le parentesi tonde in questo contesto creano una tupla
    # cioè una sequenza immutabile di elementi
    my_app.children = (my_app.children[0], # il widget del titolo precedente
                       nuova_pagina, # widget che rappresenta la nuova pagina
                       my_app.children[2]) # il widget dei credits precedente

def bottone_vai_pag2_cliccato(b):
    cambia_pagina(pagina2)

bottone_vai_pag2.on_click(bottone_vai_pag2_cliccato)

def bottone_vai_pag1_cliccato(b):
    cambia_pagina(pagina1)

bottone_vai_pag1.on_click(bottone_vai_pag1_cliccato)
display(my_app)
````

````python
%matplotlib inline
import matplotlib.pyplot as plt

with open('Alberi-Monumentali-Della-Campania.csv', encoding='utf-8', newline='') as f:
    lettore = csv.DictReader(f, delimiter=';')
    next(lettore)
    ret = []
    province = {}
    for d in lettore:
        p = d['PROVINCIA']
        if p not in province:
            province[p] = 1
        else:
            province[p] += 1

    xs = list(range(len(province)))
    nomi_province = province.keys()
    ys = [province[x] for x in nomi_province]
    plt.xticks(xs, nomi_province)
    plt.title('Alberi monumentali per provincia')
    plt.bar(xs, ys, 0.5, color='green', align='center')
    plt.show()
````
